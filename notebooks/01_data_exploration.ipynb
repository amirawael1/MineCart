```python
# Milestone 1: Data Exploration
# Amazon QA Project - Initial Analysis

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("NLP Project: Amazon Question Answering System")
print("="*60)
print("Author: Amira Wael")
print("Milestone 1: Data Analysis")

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Create sample data for initial exploration
np.random.seed(42)
n_samples = 100

data = pd.DataFrame({
    'product_id': [f'PROD_{i:04d}' for i in range(n_samples)],
    'title': [f'Amazon Product {i}' for i in range(n_samples)],
    'category': np.random.choice(['Electronics', 'Books', 'Clothing', 'Home'], n_samples),
    'price': np.random.uniform(5, 300, n_samples),
    'rating': np.random.uniform(1, 5, n_samples),
    'review_count': np.random.randint(0, 500, n_samples)
})

print("\n1. Dataset Overview:")
print(f"Shape: {data.shape}")
print(f"Columns: {list(data.columns)}")

print("\n2. First 5 rows:")
print(data.head())

print("\n3. Basic Statistics:")
print(data.describe())

print("\n4. Missing Values:")
print(data.isnull().sum())

print("\nâœ… Ready for Milestone 1 analysis!")
```
